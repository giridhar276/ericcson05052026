# Lambda Functions in Python - Simple to Complex

This notebook explains lambda functions from basic to slightly advanced examples.

## Example 1


```python
# Create a lambda function to add two numbers.
add = lambda a, b: a + b

# Call the lambda function with two values.
result = add(10, 20)

# Print the result.
print("Addition:", result)
```

## Example 2


```python
# Create a lambda function to square a number.
square = lambda number: number * number

# Call the lambda function.
result = square(6)

# Print the square result.
print("Square:", result)
```

## Example 3


```python
# Create a lambda function to check even or odd.
even_or_odd = lambda number: "Even" if number % 2 == 0 else "Odd"

# Call lambda function with number 10.
print("10 is:", even_or_odd(10))

# Call lambda function with number 7.
print("7 is:", even_or_odd(7))
```

## Example 4


```python
# Create a list of employee tuples.
employees = [("Asha", 45000), ("Ravi", 65000), ("Meena", 55000)]

# Sort employees based on salary using lambda.
employees_sorted = sorted(employees, key=lambda employee: employee[1])

# Print sorted employees.
print("Employees sorted by salary:", employees_sorted)
```

## Example 5


```python
# Create a list of dictionaries.
products = [
    {"name": "Laptop", "price": 65000},
    {"name": "Mouse", "price": 800},
    {"name": "Keyboard", "price": 1500}
]

# Sort products by price using lambda function.
sorted_products = sorted(products, key=lambda product: product["price"])

# Print sorted products.
print("Products sorted by price:", sorted_products)
```
