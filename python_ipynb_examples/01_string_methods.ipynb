# String Methods in Python

This notebook explains commonly used Python string methods with simple examples and line-by-line comments.

## Example 1


```python
# Create a string variable with extra spaces at the beginning and end.
message = "   welcome to python training   "

# Print the original string exactly as it is stored.
print("Original message:", message)

# strip() removes spaces from both beginning and end.
clean_message = message.strip()

# Print the cleaned string.
print("After strip():", clean_message)

# upper() converts all characters to uppercase.
print("Uppercase:", clean_message.upper())

# lower() converts all characters to lowercase.
print("Lowercase:", clean_message.lower())

# title() converts the first letter of each word to uppercase.
print("Title case:", clean_message.title())

# capitalize() converts only the first character to uppercase.
print("Capitalize:", clean_message.capitalize())
```

## Example 2


```python
# Create a sentence to practice searching and replacing.
sentence = "Python is easy and Python is powerful"

# count() counts how many times a substring appears.
python_count = sentence.count("Python")

# Print the count result.
print("Count of Python:", python_count)

# replace() replaces one substring with another substring.
new_sentence = sentence.replace("Python", "Java")

# Print the replaced sentence.
print("After replace():", new_sentence)

# find() returns the index position of the first matching substring.
position = sentence.find("easy")

# Print the position of the word easy.
print("Position of easy:", position)

# startswith() checks if the string starts with given text.
print("Starts with Python:", sentence.startswith("Python"))

# endswith() checks if the string ends with given text.
print("Ends with powerful:", sentence.endswith("powerful"))
```

## Example 3


```python
# Create a comma-separated string.
data = "Asha,Ravi,Meena,Arjun"

# split() breaks the string into a list based on comma.
names = data.split(",")

# Print the list created from the string.
print("Names list:", names)

# join() joins list items into a single string.
joined_names = " | ".join(names)

# Print the joined string.
print("Joined names:", joined_names)

# isalpha() checks whether all characters are alphabets.
print("Asha is alphabetic:", "Asha".isalpha())

# isdigit() checks whether all characters are digits.
print("12345 is digit:", "12345".isdigit())

# isalnum() checks whether all characters are alphabets or numbers.
print("User123 is alphanumeric:", "User123".isalnum())
```
