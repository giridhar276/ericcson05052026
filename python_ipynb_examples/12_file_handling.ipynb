# File Handling in Python

This notebook explains file handling using employee.csv with read(), readline(), readlines(), file object, write(), and writelines().

## Example 1


```python
# Open employee.csv file in read mode.
fobj = open("employee.csv", "r")

# read() reads the complete file content as one string.
content = fobj.read()

# Print the complete content.
print(content)

# Close the file after reading.
fobj.close()
```

## Example 2


```python
# Open employee.csv file in read mode.
fobj = open("employee.csv", "r")

# readline() reads only one line from the file.
first_line = fobj.readline()

# Print the first line.
print("First line:", first_line)

# readline() reads the next line from the file.
second_line = fobj.readline()

# Print the second line.
print("Second line:", second_line)

# Close the file after reading.
fobj.close()
```

## Example 3


```python
# Open employee.csv file in read mode.
fobj = open("employee.csv", "r")

# readlines() reads all lines and returns a list.
lines = fobj.readlines()

# Print the list of lines.
print(lines)

# Loop through each line from the list.
for line in lines:
    # Print each line after removing extra newline.
    print(line.strip())

# Close the file after reading.
fobj.close()
```

## Example 4


```python
# Open a new file in write mode.
fobj = open("output_employees.txt", "w")

# write() writes a single string to the file.
fobj.write("Employee Report\n")

# write() writes another single string to the file.
fobj.write("This file is created using write() method.\n")

# Close the file after writing.
fobj.close()

# Print confirmation message.
print("output_employees.txt created successfully")
```

## Example 5


```python
# Create a list of lines to write into a file.
lines = [
    "Python File Handling\n",
    "write() writes one string at a time\n",
    "writelines() writes multiple strings from a list\n"
]

# Open a new file in write mode.
fobj = open("notes.txt", "w")

# writelines() writes all strings from the list.
fobj.writelines(lines)

# Close the file after writing.
fobj.close()

# Print confirmation message.
print("notes.txt created successfully")
```
