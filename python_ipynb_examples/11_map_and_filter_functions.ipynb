# Map and Filter Functions in Python

This notebook explains map() and filter() functions with lambda and normal functions.

## Example 1


```python
# Create a list of numbers.
numbers = [1, 2, 3, 4, 5]

# Use map() with lambda to square each number.
squares = list(map(lambda number: number * number, numbers))

# Print the squared numbers.
print("Squares:", squares)
```

## Example 2


```python
# Create a list of salaries.
salaries = [30000, 45000, 60000, 75000]

# Use map() to increase each salary by 10 percent.
updated_salaries = list(map(lambda salary: salary + salary * 0.10, salaries))

# Print updated salaries.
print("Updated salaries:", updated_salaries)
```

## Example 3


```python
# Create a list of numbers.
numbers = [10, 15, 20, 25, 30, 35]

# Use filter() to select only even numbers.
even_numbers = list(filter(lambda number: number % 2 == 0, numbers))

# Print even numbers.
print("Even numbers:", even_numbers)
```

## Example 4


```python
# Create a list of employee dictionaries.
employees = [
    {"name": "Asha", "salary": 45000},
    {"name": "Ravi", "salary": 65000},
    {"name": "Meena", "salary": 55000}
]

# Use filter() to select employees whose salary is greater than 50000.
high_salary_employees = list(filter(lambda employee: employee["salary"] > 50000, employees))

# Print high salary employees.
print("High salary employees:", high_salary_employees)
```

## Example 5


```python
# Create a function to convert text into uppercase.
def convert_upper(text):
    # Return uppercase version of the text.
    return text.upper()

# Create a list of names.
names = ["asha", "ravi", "meena"]

# Use map() with a normal function.
upper_names = list(map(convert_upper, names))

# Print uppercase names.
print("Uppercase names:", upper_names)
```
