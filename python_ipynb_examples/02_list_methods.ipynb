# List Methods in Python

This notebook explains important Python list methods with examples and line-by-line comments.

## Example 1


```python
# Create a list of programming languages.
languages = ["Python", "Java", "C++"]

# Print the original list.
print("Original list:", languages)

# append() adds one item at the end of the list.
languages.append("JavaScript")

# Print the list after append.
print("After append():", languages)

# insert() adds an item at a specific index position.
languages.insert(1, "Go")

# Print the list after insert.
print("After insert():", languages)

# extend() adds multiple items from another list.
languages.extend(["Ruby", "Perl"])

# Print the list after extend.
print("After extend():", languages)
```

## Example 2


```python
# Create a list of numbers.
numbers = [10, 20, 30, 20, 40, 50]

# remove() removes the first matching value from the list.
numbers.remove(20)

# Print the list after removing first 20.
print("After remove():", numbers)

# pop() removes an item by index and returns it.
removed_item = numbers.pop(2)

# Print the removed item.
print("Removed item:", removed_item)

# Print the list after pop.
print("After pop():", numbers)

# clear() removes all elements from the list.
numbers.clear()

# Print the empty list.
print("After clear():", numbers)
```

## Example 3


```python
# Create a list of marks.
marks = [78, 92, 65, 88, 92, 70]

# count() counts how many times a value appears.
print("Count of 92:", marks.count(92))

# index() returns the first index position of a value.
print("Index of 65:", marks.index(65))

# sort() arranges the list in ascending order.
marks.sort()

# Print the sorted list.
print("Ascending order:", marks)

# reverse() reverses the current list order.
marks.reverse()

# Print the reversed list.
print("Reversed order:", marks)

# copy() creates a shallow copy of the list.
marks_copy = marks.copy()

# Print the copied list.
print("Copied list:", marks_copy)
```
