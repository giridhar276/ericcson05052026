# Loop Statements in Python

This notebook explains for loops, while loops, break, continue, and nested loops.

## Example 1


```python
# Create a list of names.
names = ["Asha", "Ravi", "Meena", "Arjun"]

# Use for loop to iterate each name.
for name in names:
    # Print each name from the list.
    print("Name:", name)
```

## Example 2


```python
# Start a counter variable with value 1.
counter = 1

# Run the while loop until counter is less than or equal to 5.
while counter <= 5:
    # Print the current counter value.
    print("Counter:", counter)

    # Increase counter value by 1 to avoid infinite loop.
    counter += 1
```

## Example 3


```python
# Use range() to generate numbers from 1 to 10.
for number in range(1, 11):
    # Check if number is equal to 5.
    if number == 5:
        # break stops the loop immediately.
        break

    # Print the number before break happens.
    print("Number:", number)
```

## Example 4


```python
# Use range() to generate numbers from 1 to 10.
for number in range(1, 11):
    # Check whether number is even.
    if number % 2 == 0:
        # continue skips the remaining code for this iteration.
        continue

    # Print only odd numbers.
    print("Odd number:", number)
```

## Example 5


```python
# Outer loop runs from 1 to 3.
for i in range(1, 4):
    # Inner loop runs from 1 to 3.
    for j in range(1, 4):
        # Print multiplication of i and j.
        print(i, "x", j, "=", i * j)
```
