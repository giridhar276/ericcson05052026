# Different Ways of Passing Arguments in Python

This notebook explains positional, keyword, default, variable-length, and keyword variable-length arguments.

## Example 1


```python
# Define a function with two parameters.
def employee_details(name, department):
    # Print employee name.
    print("Name:", name)

    # Print employee department.
    print("Department:", department)

# Pass arguments based on position.
employee_details("Asha", "HR")
```

## Example 2


```python
# Define a function with two parameters.
def employee_details(name, department):
    # Print employee name.
    print("Name:", name)

    # Print employee department.
    print("Department:", department)

# Pass arguments using parameter names.
employee_details(department="IT", name="Ravi")
```

## Example 3


```python
# Define a function with a default argument.
def greet(name, message="Welcome"):
    # Print message and name.
    print(message, name)

# Call function by passing only required argument.
greet("Meena")

# Call function by overriding default argument.
greet("Arjun", "Good morning")
```

## Example 4


```python
# Define a function using *args for variable number of positional arguments.
def add_all_numbers(*numbers):
    # Create a variable to store total.
    total = 0

    # Loop through all received numbers.
    for number in numbers:
        # Add each number to total.
        total += number

    # Return final total.
    return total

# Call function with multiple values.
result = add_all_numbers(10, 20, 30, 40)

# Print the result.
print("Total:", result)
```

## Example 5


```python
# Define a function using **kwargs for variable number of keyword arguments.
def print_profile(**details):
    # Loop through key-value pairs.
    for key, value in details.items():
        # Print each key and value.
        print(key, ":", value)

# Call function with keyword arguments.
print_profile(name="Kavya", role="Data Analyst", city="Hyderabad")
```
