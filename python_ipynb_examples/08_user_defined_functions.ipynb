# User Defined Functions in Python

This notebook explains how to create and call user-defined functions in Python.

## Example 1


```python
# Define a simple function named greet.
def greet():
    # This line runs when the function is called.
    print("Welcome to Python functions")

# Call the function.
greet()
```

## Example 2


```python
# Define a function with one parameter.
def greet_user(name):
    # Print greeting using the received name.
    print("Hello", name)

# Call the function by passing one argument.
greet_user("Asha")
```

## Example 3


```python
# Define a function that returns a value.
def add_numbers(a, b):
    # Add both numbers and store result.
    result = a + b

    # Return the result to the caller.
    return result

# Call the function and store returned value.
total = add_numbers(10, 20)

# Print the returned value.
print("Total:", total)
```

## Example 4


```python
# Define a function to calculate simple interest.
def calculate_simple_interest(principal, rate, time):
    # Apply simple interest formula.
    interest = (principal * rate * time) / 100

    # Return calculated interest.
    return interest

# Call the function with sample values.
si = calculate_simple_interest(10000, 8, 2)

# Print the simple interest.
print("Simple Interest:", si)
```
