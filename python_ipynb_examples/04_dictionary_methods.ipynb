# Dictionary Methods in Python

This notebook explains commonly used Python dictionary methods with examples and line-by-line comments.

## Example 1


```python
# Create a dictionary with employee information.
employee = {"id": 101, "name": "Asha", "department": "HR", "salary": 45000}

# Print the original dictionary.
print("Employee dictionary:", employee)

# keys() returns all keys from the dictionary.
print("Keys:", employee.keys())

# values() returns all values from the dictionary.
print("Values:", employee.values())

# items() returns key-value pairs from the dictionary.
print("Items:", employee.items())

# get() safely reads a value using a key.
print("Employee name:", employee.get("name"))

# get() returns default value if key is not found.
print("Bonus:", employee.get("bonus", "Not available"))
```

## Example 2


```python
# Create a dictionary with product details.
product = {"id": 501, "name": "Laptop", "price": 65000}

# update() adds or modifies key-value pairs.
product.update({"price": 60000, "brand": "Lenovo"})

# Print the dictionary after update.
print("After update():", product)

# pop() removes a key and returns its value.
removed_price = product.pop("price")

# Print the removed value.
print("Removed price:", removed_price)

# Print the dictionary after pop.
print("After pop():", product)

# popitem() removes the last inserted key-value pair.
last_item = product.popitem()

# Print the removed key-value pair.
print("Removed last item:", last_item)

# Print the dictionary after popitem.
print("After popitem():", product)
```

## Example 3


```python
# Create a dictionary with student marks.
student = {"name": "Ravi", "math": 85, "science": 90}

# setdefault() returns value if key exists.
name = student.setdefault("name", "Unknown")

# Print existing value.
print("Name:", name)

# setdefault() adds the key if it does not exist.
student.setdefault("city", "Hyderabad")

# Print the dictionary after setdefault.
print("After setdefault():", student)

# copy() creates a shallow copy of dictionary.
student_copy = student.copy()

# Print the copied dictionary.
print("Copied dictionary:", student_copy)

# clear() removes all items from dictionary.
student.clear()

# Print the empty dictionary.
print("After clear():", student)
```
