# Conditions in Python

This notebook explains if, if-else, if-elif-else, nested if, and conditional expressions.

## Example 1


```python
# Store the marks of a student.
marks = 78

# Check whether marks are greater than or equal to 35.
if marks >= 35:
    # This block runs only when condition is True.
    print("Student passed")
```

## Example 2


```python
# Store the age of a person.
age = 17

# Check whether age is greater than or equal to 18.
if age >= 18:
    # This block runs if the person is eligible.
    print("Eligible to vote")
else:
    # This block runs if the person is not eligible.
    print("Not eligible to vote")
```

## Example 3


```python
# Store marks for grading.
marks = 86

# Check for grade A.
if marks >= 90:
    # Print grade A.
    print("Grade A")
elif marks >= 75:
    # Print grade B.
    print("Grade B")
elif marks >= 60:
    # Print grade C.
    print("Grade C")
else:
    # Print fail if none of the above conditions are true.
    print("Needs improvement")
```

## Example 4


```python
# Store username and password.
username = "admin"

# Store password.
password = "python123"

# First check the username.
if username == "admin":
    # If username is correct, check password.
    if password == "python123":
        # This runs when both username and password are correct.
        print("Login successful")
    else:
        # This runs when password is incorrect.
        print("Invalid password")
else:
    # This runs when username is incorrect.
    print("Invalid username")
```
