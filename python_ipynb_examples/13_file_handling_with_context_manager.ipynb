# File Handling with Context Manager in Python

This notebook explains the recommended way of file handling using the with statement/context manager.

## Example 1


```python
# Open employee.csv using context manager.
with open("employee.csv", "r") as fobj:
    # read() reads the complete file content.
    content = fobj.read()

    # Print the file content.
    print(content)

# File is automatically closed after the with block.
```

## Example 2


```python
# Open employee.csv using context manager.
with open("employee.csv", "r") as fobj:
    # readlines() reads all lines into a list.
    lines = fobj.readlines()

# Loop through each line outside the with block.
for line in lines:
    # Print each line after removing newline.
    print(line.strip())
```

## Example 3


```python
# Create a list of report lines.
report_lines = [
    "Employee Salary Report\n",
    "Asha - 45000\n",
    "Ravi - 65000\n",
    "Meena - 55000\n"
]

# Open salary_report.txt using context manager in write mode.
with open("salary_report.txt", "w") as fobj:
    # writelines() writes multiple lines into the file.
    fobj.writelines(report_lines)

# Print confirmation message.
print("salary_report.txt created successfully")
```

## Example 4


```python
# Open employee.csv using context manager.
with open("employee.csv", "r") as fobj:
    # Skip the header line using readline().
    header = fobj.readline()

    # Read remaining employee records.
    records = fobj.readlines()

# Print the header.
print("Header:", header.strip())

# Loop through records.
for record in records:
    # Split each CSV row by comma.
    fields = record.strip().split(",")

    # Store employee name from index 1.
    employee_name = fields[1]

    # Store salary from index 3 and convert to integer.
    salary = int(fields[3])

    # Print employees whose salary is greater than 50000.
    if salary > 50000:
        # Print selected employee details.
        print(employee_name, "has salary greater than 50000")
```
