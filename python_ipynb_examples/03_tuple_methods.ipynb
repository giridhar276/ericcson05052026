# Tuple Methods in Python

This notebook explains tuple basics and the two main tuple methods: count() and index().

## Example 1


```python
# Create a tuple of city names.
cities = ("Hyderabad", "Chennai", "Bengaluru", "Hyderabad", "Pune")

# Print the complete tuple.
print("Cities tuple:", cities)

# count() counts how many times a value appears in the tuple.
hyderabad_count = cities.count("Hyderabad")

# Print the count result.
print("Count of Hyderabad:", hyderabad_count)

# index() returns the first index position of a value.
chennai_index = cities.index("Chennai")

# Print the index result.
print("Index of Chennai:", chennai_index)
```

## Example 2


```python
# Create a tuple of employee details.
employee = (101, "Asha", "HR", 45000)

# Access the first item using index 0.
print("Employee ID:", employee[0])

# Access the second item using index 1.
print("Employee Name:", employee[1])

# Access the third item using index 2.
print("Department:", employee[2])

# Access the fourth item using index 3.
print("Salary:", employee[3])

# len() returns the total number of items in the tuple.
print("Total items:", len(employee))
```

## Example 3


```python
# Create a tuple containing numbers.
numbers = (10, 20, 30, 40, 50)

# Use slicing to get first three items.
first_three = numbers[0:3]

# Print the sliced tuple.
print("First three numbers:", first_three)

# Use slicing to get last two items.
last_two = numbers[-2:]

# Print the last two items.
print("Last two numbers:", last_two)

# Use a for loop to iterate through tuple items.
for number in numbers:
    # Print each number from the tuple.
    print("Number:", number)
```
