# Set Methods in Python

This notebook explains important Python set methods with simple examples and comments.

## Example 1


```python
# Create a set of skills.
skills = {"Python", "SQL", "Excel"}

# Print the original set.
print("Original set:", skills)

# add() adds one item to the set.
skills.add("Power BI")

# Print the set after add.
print("After add():", skills)

# update() adds multiple items to the set.
skills.update(["Machine Learning", "Python", "Tableau"])

# Print the set after update.
print("After update():", skills)

# Notice that duplicate Python is not added again because sets store unique values.
print("Final skills:", skills)
```

## Example 2


```python
# Create a set of numbers.
numbers = {10, 20, 30, 40, 50}

# remove() removes an item but gives error if item does not exist.
numbers.remove(20)

# Print the set after remove.
print("After remove():", numbers)

# discard() removes an item and does not give error if item does not exist.
numbers.discard(100)

# Print the set after discard.
print("After discard():", numbers)

# pop() removes a random item from the set.
removed_item = numbers.pop()

# Print the removed item.
print("Removed item:", removed_item)

# Print the set after pop.
print("After pop():", numbers)
```

## Example 3


```python
# Create two sets.
python_students = {"Asha", "Ravi", "Meena", "Arjun"}

# Create another set.
sql_students = {"Ravi", "Kavya", "Arjun", "Neha"}

# union() returns all unique items from both sets.
print("Union:", python_students.union(sql_students))

# intersection() returns common items from both sets.
print("Intersection:", python_students.intersection(sql_students))

# difference() returns items present in first set but not in second set.
print("Only Python students:", python_students.difference(sql_students))

# symmetric_difference() returns items not common in both sets.
print("Different students:", python_students.symmetric_difference(sql_students))
```
