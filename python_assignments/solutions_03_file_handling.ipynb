## Solution 1: Create High Salary Report


```python
# Open employee.csv in read mode.
input_file = open("employee.csv", "r")

# Read all lines from the file.
lines = input_file.readlines()

# Open output file in write mode.
output_file = open("high_salary_report.txt", "w")

# Write header to output file.
output_file.write("emp_id,name,department,salary,city,experience\n")

# Loop through all employee records except header.
for line in lines[1:]:
    # Split line by comma.
    fields = line.strip().split(",")

    # Convert salary value into integer.
    salary = int(fields[3])

    # Check if salary is greater than 60000.
    if salary > 60000:
        # Write matching employee record to output file.
        output_file.write(line)

# Close input file.
input_file.close()

# Close output file.
output_file.close()

# Print success message.
print("high_salary_report.txt created successfully")
```

## Solution 2: Filter IT Employees using Context Manager


```python
# Open employee.csv using context manager.
with open("employee.csv", "r") as input_file:
    # Read all lines from employee.csv.
    lines = input_file.readlines()

# Open output file using context manager.
with open("it_employees.txt", "w") as output_file:
    # Write header line.
    output_file.write(lines[0])

    # Loop through employee records except header.
    for line in lines[1:]:
        # Split line into fields.
        fields = line.strip().split(",")

        # Store department value.
        department = fields[2]

        # Check if department is IT.
        if department == "IT":
            # Write IT employee record into output file.
            output_file.write(line)

# Print success message.
print("it_employees.txt created successfully")
```

## Solution 3: Calculate Total Sales using Context Manager


```python
# Create total sales variable.
total_sales = 0

# Open sales.csv using context manager.
with open("sales.csv", "r") as file:
    # Skip the header line.
    header = file.readline()

    # Loop through each remaining line.
    for line in file:
        # Split CSV row by comma.
        fields = line.strip().split(",")

        # Convert quantity to integer.
        quantity = int(fields[4])

        # Convert price to integer.
        price = int(fields[5])

        # Calculate order total and add to total sales.
        total_sales += quantity * price

# Print total sales.
print("Total Sales:", total_sales)
```
