## Solution 1: Clean Customer Names


```python
# Create a list of customer names with spaces and inconsistent cases.
customer_names = ["   asha rao   ", "RAVI kumar", "  meena SHARMA  ", "arjun reddy"]

# Create an empty list to store cleaned names.
cleaned_names = []

# Loop through each customer name.
for name in customer_names:
    # Remove extra spaces using strip() and convert to title case using title().
    cleaned_name = name.strip().title()

    # Add the cleaned name to the cleaned_names list.
    cleaned_names.append(cleaned_name)

# Print the final cleaned names list.
print(cleaned_names)
```

## Solution 2: Top 3 Scores


```python
# Create a list of quiz scores.
scores = [78, 92, 65, 88, 95, 70, 82]

# Sort the scores in descending order.
scores.sort(reverse=True)

# Select top 3 scores using slicing.
top_three_scores = scores[:3]

# Print the top 3 scores.
print("Top 3 Scores:", top_three_scores)
```

## Solution 3: Department-Wise Employee Count


```python
# Create a list of departments.
departments = ["IT", "HR", "IT", "Finance", "HR", "IT", "Sales"]

# Create an empty dictionary to store department counts.
department_count = {}

# Loop through each department.
for department in departments:
    # Get existing count using get(); if missing, use 0, then add 1.
    department_count[department] = department_count.get(department, 0) + 1

# Print final department count.
print(department_count)
```
