## Solution 1: Salary Hike Function


```python
# Define a function to calculate salary after hike.
def calculate_hike(salary, hike_percentage):
    # Calculate hike amount.
    hike_amount = salary * hike_percentage / 100

    # Calculate updated salary.
    updated_salary = salary + hike_amount

    # Return the updated salary.
    return updated_salary

# Call the function with sample values.
new_salary = calculate_hike(50000, 10)

# Print updated salary.
print("Updated Salary:", new_salary)
```

## Solution 2: Sort Employees by Salary using Lambda


```python
# Create a list of employee tuples.
employees = [("Asha", 45000), ("Ravi", 65000), ("Meena", 55000)]

# Sort employees based on second value, which is salary.
sorted_employees = sorted(employees, key=lambda employee: employee[1])

# Print sorted employees.
print(sorted_employees)
```

## Solution 3: Apply Salary Hike using Map


```python
# Create a list of salaries.
salaries = [45000, 55000, 65000, 70000]

# Use map() and lambda to increase every salary by 10 percent.
updated_salaries = list(map(lambda salary: salary + salary * 0.10, salaries))

# Print updated salaries.
print(updated_salaries)
```
